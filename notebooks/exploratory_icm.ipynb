```python
import json

results_file = r"C:\Users\soswo\OneDrive\Desktop\projects\Unsupervised-Elicitation\icm_results\truthful_qa_distilgpt2_icm_20251119_095638.jsonl"

labeled_examples = []
with open(results_file, "r", encoding="utf-8") as f:
    for line in f:
        labeled_examples.append(json.loads(line))

print(f"Loaded {len(labeled_examples)} examples")
```

```python
from collections import Counter

labels = [ex["label"] for ex in labeled_examples]
label_counts = Counter(labels)

print(label_counts)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(label_counts.keys(), label_counts.values(), color=['green', 'red'])
plt.title("ICM Labeled Examples Distribution")
plt.xlabel("Label")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(label_counts.values(), labels=label_counts.keys(), autopct='%1.1f%%', colors=['green','red'])
plt.title("ICM Labeled Examples Distribution")
plt.show()
```
